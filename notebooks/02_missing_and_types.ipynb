## Purpose

## 1. Reload the raw dataset

```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

url = "https://raw.githubusercontent.com/hiroki-kaifu/messy-employee-data-cleaning/refs/heads/main/data/raw/Messy_Employee_dataset.csv"
df = pd.read_csv(url)
```

## 2. Review missing values again with emphasis on key columns

In this section, I re-check missingness with a focus on columns that are likely to affect downstream analysis.
The most important candidates are columns such as `Age` and `Salary`, where missing values may require explicit treatment decisions.

```python
missing_counts = df.isna().sum()
missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)
missing_counts
```

Missing values are concentrated in a small number of columns, especially in key analytical fields such as Age and Salary.
These columns will require explicit treatment decisions in the next steps.

## 3. Compare missingness in counts and percentages

Raw counts show the scale of the issue, while percentages make it easier to judge how serious the missingness is relative to the full dataset.
This helps determine whether simple imputation is reasonable.

```python
missing_pct = (df.isna().mean() * 100).sort_values(ascending=False).round(2)
missing_pct
```
